# Clustering optimization

```python
def calc_enrichment_temp(genelist,limit = 10000,direction="g"):  # g for greater, le for lower or equal
    geneset = set(genelist)
    results = pd.DataFrame(columns=['Total','inCluster','ratio','pvalue'])
    for term in gobp2genes:
        if len(gobp2genes[term]) == 0:
            continue
        if direction == "g" and len(gobp2genes[term]) > limit:
            continue
        if direction == "le" and len(gobp2genes[term]) <= limit:
            continue
        #[[A&B, A&~B],[~A&B, ~A&~B]]
        
        a1 = len(geneset & set(gobp2genes_all[term].keys()))
        a2 = len(geneset - set(gobp2genes_all[term].keys()))
        b1 = len(set(gobp2genes_all[term].keys())-geneset)
        b2 = 19032 - a1 - a2 - b1
        cont_table = [[a1,a2],[b1,b2]] 
        odd,pvalue = stats.fisher_exact(cont_table,alternative='greater')
        results.loc[term] = [len(gobp2genes[term]),a1,float(a1)/float(len(gobp2genes[term])),pvalue]

    #calc bh
    count=0
    bhpvalue = dict()
    results = results.sort_values('pvalue') # presorted
    bh =  statsmodels.stats.multitest.multipletests(results['pvalue'],method='fdr_bh',is_sorted=True)

    results['fdr'] = bh[1]
    
    return results 
```

```python
# define the number of clusters regarding functional enrichment.
# conditional LLS


result_lls_gobp = dict()
result_fc_gobp =dict()

gsnet = dict()
for desc in gobp2genes_all:
    if len(gobp2genes_all[desc].keys()) > 50:  # skip terms above 50
        continue
    genes = intersect1d(network_df.index,list(gobp2genes_all[desc].keys()))
    for i in range(len(genes)):
        ga = genes[i]
        if ga not in gsnet:
            gsnet[ga] = dict()
        for j in range(i+1,len(genes)):
            gb = genes[j]
            if gb not in gsnet:
                gsnet[gb] = dict()
            gsnet[ga][gb] = True
            gsnet[gb][ga] = True
tp = 0
for ga in gsnet:
    tp += len(gsnet[ga])
tp = int(tp/2)
tn = ( len(network_df.index) * (len(network_df.index)-1) / 2 ) - tp

print (tp,tn)

for number in range(4,41,1):
    clusters = scipy.cluster.hierarchy.fcluster(result.dendrogram_col.linkage, number, criterion='maxclust')
    cluster2genes = dict()
    for c in unique(clusters):
        cluster2genes[c] = list()
    for i in range(len(network_df.index)):
        cluster2genes[clusters[i]].append(network_df.index[i])
        
    sumpvalues = 0.0

    p = 0
    n = 0
        
    for c in cluster2genes:
        for i in range(len(cluster2genes[c])):
            ga = cluster2genes[c][i]
            for j in range(i+1,len(cluster2genes[c])):
                gb = cluster2genes[c][j]
                if ga in gsnet and gb in gsnet[ga]:
                    p += 1
                else:
                    n += 1
    foldchange =  (float(p)/float(p+n))/(float(tp)/float(tp+tn))
    lls = log( (float(p)/float(n))/(float(tp)/float(tn)) )/ log(2)
    #print (number,p,n,lls,foldchange)
    result_lls_gobp[number] = lls 
    result_fc_gobp[number] = foldchange
```
