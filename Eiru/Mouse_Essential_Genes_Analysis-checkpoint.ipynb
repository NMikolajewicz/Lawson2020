```python
#load MGI homolog info
firstcheck = True
homologenes = dict()
human2mouse = dict() #xref
human2mouse_duplicated = dict() # xref of duplicated genes(human1 -> mouse 2~3)
with open('HOM_AllOrganism.rpt','r') as fp:
    for line in fp:
        if firstcheck == True:
            firstcheck = False
            continue
        linearray = line.rstrip().split("\t")
        homologeneid = int(linearray[0])
        taxon = int(linearray[2])
        symbol = linearray[3]
        entrezid = int(linearray[4])
        if taxon in [10090,9606]: # Mouse(10090) and Human(9606)
            if homologeneid not in homologenes:
                homologenes[homologeneid] = dict()
            if taxon not in homologenes[homologeneid]:
                homologenes[homologeneid][taxon] = list()
            homologenes[homologeneid][taxon].append((entrezid,symbol))

for i in homologenes:
    if len(homologenes[i]) == 2:
        if len(homologenes[i][9606])==1 and  len(homologenes[i][10090])==1: # only 1 to 1 
            for (hid,hsym) in homologenes[i][9606]:
                if hid in entrez2symbol.index:
                    human2mouse[hid] = homologenes[i][10090]  # only in ncbi genes
                else:
                    pass
                
        elif len(homologenes[i][10090])>1:
            for (hid,hsym) in homologenes[i][9606]:
                if hid in entrez2symbol.index:
                    human2mouse_duplicated[hid] = homologenes[i][10090]  # only in ncbi genes
                else:
                    pass
            
```

```python
# Convert Human Ref to Mouse Ref. Discard non-CCDS genes

mouse_ceg_entrez = list()
mouse_neg_entrez = list()

noortholog_ceg = list()  # no orthologship
noortholog_neg = list()
duportholog_ceg = list()  # there is ortholog but duplicated
duportholog_neg = list()
for entrezid in human_ceg_entrez:
    if entrezid not in human2mouse:
        if entrezid in human2mouse_duplicated:
            duportholog_ceg.append((entrezid,entrez2symbol[entrezid]))
        else:
            noortholog_ceg.append((entrezid,entrez2symbol[entrezid]))
        #nacount+=1
        #print entrezid, entrez2symbol[entrezid]
    else:
        if len(human2mouse[entrezid]) >= 2:
            print ("%d, %s is duplicated" % (entrezid,entrez2symbol[entrezid]))
            for (mid,msym) in human2mouse[entrezid]:
                if mid in ccds_mouse['gene_id'].values:
                    print (' Mouse:',mid,msym)
                else:
                    print( " Mouse: Not in CCDS",mid,msym)
        for (mid,msym) in human2mouse[entrezid]:
            if mid in ccds_mouse['gene_id'].values:
                mouse_ceg_entrez.append(mid)
            
for entrezid in human_neg_entrez:
    if entrezid not in human2mouse:
        if entrezid in human2mouse_duplicated:
            duportholog_neg.append((entrezid,entrez2symbol[entrezid]))
        else:
            noortholog_neg.append((entrezid,entrez2symbol[entrezid]))
        #nacount+=1
        #print entrezid, entrez2symbol[entrezid]
    else:
        if len(human2mouse[entrezid]) >= 2:
            print ("%d, %s is duplicated  " % (entrezid,entrez2symbol[entrezid]))
            for (mid,msym) in human2mouse[entrezid]:
                if mid in ccds_mouse['gene_id'].values:
                    print (' Mouse:',mid,msym)
                else:
                    print (" Mouse: Not in CCDS",mid,msym)
        for (mid,msym) in human2mouse[entrezid]:
            if mid in ccds_mouse['gene_id'].values:
                mouse_neg_entrez.append(mid)
mouse_ceg_entrez = list(set(mouse_ceg_entrez))      # rm duplicates
mouse_neg_entrez = list(set(mouse_neg_entrez))      # rm duplicates
print (len(mouse_ceg_entrez))            
print (len(mouse_neg_entrez))
```

```python
def quantileNormalize(df_input):
    df = df_input.copy()
    #compute rank
    dic = {}
    for col in df:
        dic[col] = df[col].sort_values(na_position='first').values
    sorted_df = pd.DataFrame(dic)
    #rank = sorted_df.mean(axis = 1).tolist()
    rank = sorted_df.median(axis = 1).tolist()
    #sort
    for col in df:
        # compute percentile rank [0,1] for each score in column 
        t = df[col].rank( pct=True, method='max' ).values
        # replace percentile values in column with quantile normalized score
        # retrieve q_norm score using calling rank with percentile value
        df[col] = [ np.nanpercentile( rank, i*100 ) if ~np.isnan(i) else np.nan for i in t ]
    return df
```

```python
## GENCODE M18 parsing

gene2transcript = dict()
transcript2exon = dict()
transcript2type = dict()
transcript2CDS = dict()
transcriptlength = dict()
gene2location = dict()
with open('gencode.vM18.annotation.gtf','r') as fp:
    for line in fp:
        linearray = line.rstrip().split("\t")
        #print linearray  # ['chr1', 'HAVANA', 'gene', '3073253', '3074322', '.', '+', '.', 'gene_id "ENSMUSG00000102693.1"; gene_type "TEC"; gene_name "RP23-271O17.1"; level 2; havana_gene "OTTMUSG00000049935.1";']
        chrom = linearray[0]
        featuretype = linearray[2]
        start = int(linearray[3])
        end = int(linearray[4])
        strand = linearray[6]
        tags = [x.strip().replace('"','').split(' ') for x in linearray[8].split(";")]
        
        if featuretype=='exon':
            gene=""
            transcript=""
            #genetype = ""
            for tag in tags:
                if tag[0] == 'gene_id':
                    gene=tag[1].split(".")[0]
                elif tag[0] == 'transcript_id':
                    transcript = tag[1]
                if tag[0] == 'tag' and 'appris_principal' in tag[1]:
                    transcript2type[transcript] = tag[1]
                #elif tag[0] == 'tag':
            
            if gene not in gene2transcript:
                gene2transcript[gene]=set()
            gene2transcript[gene].add(transcript)
            if transcript not in transcript2exon:
                transcript2exon[transcript] = 0
                transcriptlength[transcript]=0
            transcript2exon[transcript] += 1
            transcriptlength[transcript] += abs(end-start)+1
        if featuretype=='CDS':
            gene=""
            transcript=""
            #genetype = ""
            for tag in tags:
                if tag[0] == 'gene_id':
                    gene=tag[1].split(".")[0]
                elif tag[0] == 'transcript_id':
                    transcript = tag[1]

            if transcript not in transcript2CDS:
                transcript2CDS[transcript] = 0
            transcript2CDS[transcript] += abs(end-start)+1
            if gene not in gene2location or abs(gene2location[gene][1] - gene2location[gene][2]) < abs(start-end): # max length for loaction
                gene2location[gene] = (chrom,start,end)
        
        #break
```
